```python
# Check for the first word
for w in words[:1]:
    for char1,char2 in zip(w,w[1:]):
        print(char1, char2)
```

**Explanation of inner loop:  -> for char1,char2 in zip(w,w[1:]): <-** 

**w** as an Iterator: This iterates over the characters of the word. For example, if w is the word "Hello", it iterates over 'H', 'e', 'l', 'l', 'o'.

w[1:] as a Second Iterator: This starts from the second character of the word and iterates over the rest of the characters. Continuing with the "Hello" example, w[1:] would iterate over 'e', 'l', 'l', 'o'.

When you zip these two iterators, you get pairs of adjacent characters. The zip() function essentially combines corresponding elements from these two sequences into pairs. So for "Hello", you would get the pairs ('H', 'e'), ('e', 'l'), ('l', 'l'), ('l', 'o'). Each pair contains adjacent characters from the word.

```python
# implement a Start and an End bigram
for w in words[:1]:
    chars = ['<S>'] + list(w) + ['<E>'] # NEW
    for char1,char2 in zip(chars,chars[1:]):
        print(char1, char2)
```

**stoi = {s:i for i,s in enumerate(chars)}:** This line creates a dictionary that maps each character to a unique integer.ist.

+ enumerate(chars): This iterates over chars, providing a pair of values for each character: its index (i) and the character itself (s) 

+ {s:i for i,s in ...}: This is a dictionary comprehension that creates key-value pairs where each key is a character (s) and each value is the corresponding index (i).

The result is a dictionary stoi (short for "string to integer"), where each character from chars is a key, and its value is the index of that character in the sorted list.

## Step 1: Create the training set of all the bigrams (x,y)

## Step 4: Feeed Forward propagation: Log-counts -> counts -> probs

```python
# FF-Path (1 layer NO BIAS)
logits = xenc @ W # log-counts

# SOFTMAX - Normalization
counts= logits.exp() # we do EXPONENTITION so that it looks like counts (EQUIVALENT to N)
probs = counts / counts.sum(dim = 1, keepdims=True)

probs
```

## Step 6: WRAP UP ALL

```python
# create the dataset
xs,ys = [],[] # xs-inputs, ys-targets
for w in words: # ALL WORDS 
    chars = ['.'] + list(w) + ['.']
    for char1,char2 in zip(chars,chars[1:]):
        ix1 = stoi[char1]
        ix2 = stoi[char2]
        
        xs.append(ix1)
        ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)

num = xs.nelement()
print('number of examples: ',num)

# initialize the 'network'
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27,27),generator=g,requires_grad=True)
```

```python
# Gradient descenct
for k in range (200):

    # FORWARD PASS
    xenc = F.one_hot(xs,num_classes=27).float()
    logits = xenc @ W
    counts= logits.exp()
    probs = counts / counts.sum(dim = 1, keepdims=True)
    loss = -probs[torch.arange(num), ys].log().mean()  + 0.01*(W**2).mean() # <- adding a Regulariziation (0.01*(W**2))
    print('Loss:',loss.item())

    # BACKWARD PASS
    W.grad = None
    loss.backward() 

    # UPDATE W (learning rate: 50)
    W.data += -50 * W.grad
```

## SAMPLE FROM THE MODEL

```python
#Loop
g = torch.Generator().manual_seed(2147483647)

for i in range(5):

    ix = 0
    out = []

    while True:

        xenc = F.one_hot(torch.tensor([ix]),num_classes=27).float()
        logits = xenc @ W
        counts= logits.exp()
        p = counts / counts.sum(dim = 1, keepdims=True)
        
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
